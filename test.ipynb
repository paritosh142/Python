```python
import cv2
import mediapipe as mp
import numpy as np
```

```python
mp_face_detection = mp.solutions.face_detection 
mp_drawing = mp.solutions.drawing_utils
```

```python
front_image = cv2.imread('face/front_face.jpg')
```

```python
face_detection = mp_face_detection.FaceDetection(min_detection_confidence=0.5)
```

```python
front_image_rgb = cv2.cvtColor(front_image, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(front_image_rgb)
```

```python
face_detection = mp_face_detection.FaceDetection(min_detection_confidence=0.5)
results = face_detection.process(front_image_rgb)   

if results.detections:
    for detection in results.detections:
        mp_drawing.draw_detection(front_image, detection)

        bboxC = detection.location_data.relative_bounding_box
        ih, iw, _ = front_image.shape
        x, y, w, h = int(bboxC.xmin * iw), int(bboxC.ymin * ih), int(bboxC.width * iw), int(bboxC.height * ih)

        face = front_image_rgb[y:y+h, x:x+w]

        face_resized = cv2.resize(face, (160 , 160))
        
        plt.axis('off')
        plt.imshow(face_resized)
        

 
```

```python
import tensorflow as tf
```

```python
from keras_facenet import FaceNet
```

```python
embedder = FaceNet()    
```

```python
embeddings = embedder.embeddings([face_resized])
```

```python
print(embeddings)   
```
